```python
import datajoint as dj
import pandas as pd
import numpy as np

acquisition = dj.create_virtual_module('acquisition', 'u19_acquisition')
behavior = dj.create_virtual_module('behavior', 'u19_behavior')
```

## Fetch all sessions from a subject

```python
#key = dict()
#key['subject_fullname'] = 'ms81_M005'
key = "subject_fullname like 'emdia%'" 
#key['session_date'] = '2021-05-05'

print(key)


all_sessions_df = pd.DataFrame((acquisition.Session & key).fetch(as_dict=True))
all_sessions_df
```

## All trials DataFrame

```python
blocktable = behavior.TowersBlock() * behavior.TowersBlock.Trial()
#Remove position, velocity fields because they take a lot to load
all_fields = pd.DataFrame.from_dict(blocktable.heading.attributes, orient='index')
sel_fields = all_fields.loc[~all_fields['name'].isin(['position', 'velocity', 'sensor_dots', 'collision', 'trial_time']), :]
sel_fields = sel_fields['name'].tolist()


all_trials_df = pd.DataFrame((blocktable & key).fetch(*sel_fields, as_dict=True))
all_trials_df
```

```python
all_trials_df['towers_left']  = all_trials_df.apply(lambda x: np.sum(x['cue_presence_left'][0]),axis=1)
all_trials_df['towers_right']  = all_trials_df.apply(lambda x: np.sum(x['cue_presence_right'][0]),axis=1)
all_trials_df['stimulus']  =  all_trials_df['towers_left'] - all_trials_df['towers_right']
all_trials_df['choice']  = all_trials_df['choice'].replace({'L': 0, 'R': 1, 'nil': -1})
all_trials_df['trial_type']  = all_trials_df['trial_type'].replace({'L': 0, 'R': 1, 'nil': -1})

all_trials_df  = all_trials_df.loc[(all_trials_df['main_level'] == 11) & (all_trials_df['level'] == 11), :]
all_trials_df = all_trials_df.reset_index(drop=True)
```
